```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def init_random(X,K):

  randidx = np.random.permutation(X.shape[0])

  centroids = X[randidx[:K]]

  return centroids
```

```python
def find_closest_centroids(X, centroids):

    K = centroids.shape[0]

    index = np.zeros(X.shape[0], dtype=int)

    for i in range(X.shape[0]):
          dis = []
          for j in range(centroids.shape[0]):
              norm_ij = np.linalg.norm(X[i] - centroids[j])
              dis.append(norm_ij)

          index[i] = np.argmin(dis)


    return index
```

```python
def compute_centroids(X, index, K):


    m, n = X.shape

    centroids = np.zeros((K, n))

    for k in range(K):
          points = X[index == k]
          centroids[k] = np.mean(points, axis = 0)


    return centroids
```

```python
def run_kMeans(X, initial_centroids, max_iters=10):


    m, n = X.shape
    K = initial_centroids.shape[0]
    centroids = initial_centroids
    previous_centroids = centroids
    index = np.zeros(m)

    for i in range(max_iters):

        print("\nK-Means iteration %d/%d" % (i, max_iters-1))

        index = find_closest_centroids(X, centroids)

        print('\nCentroids in ',i+1, 'iteration:\n', centroids, '\n')
        plt.scatter(X[:,0],X[:,1], c = index)
        plt.scatter(centroids[:,0],centroids[:,1], c = 'red', s = 200)
        plt.show()
        centroids = compute_centroids(X, index, K)

    return centroids, index
```
